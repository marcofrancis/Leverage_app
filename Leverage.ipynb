# Restaking and Leverage
We want to investigate the relationship between restaking and leverage. In particular we want to se if we can replicate the exposure of a restaking strategy withthe use of a leveraged portfolio.

## Restaking
Let's suppose that we have a wealth $w$ in ETH to restake in two AVS. In this setting let's indicate with $\phi_1$ and $\phi_2$ the amount of ETH we invest in the first and second AVS.

Each AVS will promise a certain reward, and we will face a certain risk and cost when we restake our ETH with them. To model this we will simply assume that the payoff from restaking the amount $w\phi_1$ in the first AVS is given by $w\phi_1(1+r_1)$ and similarly for the second AVS we have a payoff of $w\phi_2(1+r_2)$.

The interesting bit about restaking is that we can increase our capital efficiency and get rewards from multiple sources for the same amount of capital invested. More precisely we can model the total reward from restaking as:

$$
w\left[\phi_{1}\left(1+r_{1}\right)+\phi_{2}\left(1+r_{2}\right)+(1-\phi_{1}-\phi_{2})(1+r_{1}+r_{2})\right]=w\left[1+r_{1}\left(1-\phi_{2}\right)+r_{2}(1-\phi_{1})\right]
$$
where $0\leq \phi_1 + \phi_2 \leq 1$.

## Leveraged Portfolios
Another, more classical, way to invest our capital, could be to construct a leveraged position in the two AVSs. In this case we will borrow at the risk-free rate $r_f$ and invest in the two AVSs. The payoff from this strategy will be given by:

$$
w\left[\varphi_{1}\left(1+r_{1}\right)+\varphi_{2}\left(1+r_{2}\right)+(1-\varphi_{1}-\varphi_{2})(1+r_{f})\right]=w\left[1+\varphi_{1}r_{1}+\varphi_{2}r_{2}+\left(1-\varphi_{1}-\varphi_{2}\right)r_{f}\right]
$$
where $0\leq \varphi_1,\varphi_2 \leq 1$. Notice that this means that we can borrow up to 100% of our wealth at the risk-free rate to invest in the two AVSs.
